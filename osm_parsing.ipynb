```python
from tqdm import tqdm
from time import sleep
import osmapi as osm
import pandas as pd

api = osm.OsmApi() 

ans_error = []

df_osm = pd.DataFrame([], columns=['id', 'timestamp', 'tag', 'lat', 'lon'])

a = pd.read_csv('node_ids.csv')['0'].tolist()

for i in tqdm(range(len(a))):
    try:
        node = api.NodeGet(a[i])
        time = node.get('timestamp')
        tag = node.get('tag')
        lat = node.get('lat')
        lon = node.get('lon')

        df_osm.loc[len(df_osm)] = [a[i], time, tag, lat, lon]
    except:
        pass
```
